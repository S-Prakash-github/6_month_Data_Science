
#### Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

#### 1.Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
### interpolate the data using linear interpolation
import matplotlib.pyplot as plt

## to create datapoints inside the linear patter we will use interpolation
plt.scatter(x,y)
```

```python
## creating x values
x_new = np.linspace(1,5,10)
y_inter = np.interp(x_new , x, y) ## interpolating y values
```

```python
print(y_inter)
```

```python
## all points are created linearly
plt.scatter(x_new,y_inter)
```

#### Cubic interpolation with scipy

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
from scipy.interpolate import interp1d
```

```python
## this will create cubic interpolation function
f = interp1d(x,y,kind= 'cubic')
```

```python
x_new = np.linspace(1,5,10)
y_inter = f(x_new) 
print(y_inter)
```

```python
## without interpolation
plt.scatter(x,y)
```

```python
## after interpolation
plt.scatter(x_new , y_inter)
```

#### Polynomial Interpolation

```python
## data 
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
## polynomial function
p = np.polyfit(x,y,2) # specify the degree value 
```

```python
x_new = np.linspace(1,5,10)
y_inter = np.polyval(p,x_new )
```

```python
print(y_inter , x_new)
```

```python
plt.scatter(x_new , y_inter)
```

```python
p = np.polyfit(x,y,3) 
x_new = np.linspace(1,5,10)
y_inter = np.polyval(p,x_new )
plt.scatter(x_new , y_inter)
```
